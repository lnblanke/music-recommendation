```python
import pandas as pd
recset = pd.read_csv("data/recom_songs.csv")
recset["track_genre"].unique()
```

```python
recset = recset.drop(recset.columns[0], axis = 1).drop_duplicates()
recset
```

```python
genre_dict = {0: 'blues',
 1: 'classical',
 2: 'country',
 3: 'disco',
 4: 'hip-hop',
 5: 'jazz',
 6: 'metal',
 7: 'pop',
 8: 'reggae',
 9: 'rock'}
```

```python
# recommendation based on user's upload
import torch
import torch.nn as nn
import numpy as np
from sklearn.preprocessing import StandardScaler
from joblib import load

# input parameter "uplaod" is the path of csv file created by function ProcessUploadMusic
def upload_recommend(upload):
    model = nn.Sequential(
        nn.Dropout(0.2),
        nn.Linear(57, 512),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(512, 256),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(128, 64),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(64, 32),
        nn.ReLU(),
        nn.Dropout(0.2),
        nn.Linear(32, 10),
    )
    
    model.load_state_dict(torch.load("weights"))
    model.eval()
    df = pd.read_csv(upload).tail(1)
    df = df.drop(df.columns[[0,1,-1]], axis = 1)
    scaler = load(open("scaler.pkl", "rb"))
    df = scaler.transform(df)
    tensor_data = torch.tensor(df, dtype=torch.float32)
    with torch.no_grad():
        y_pred = model(tensor_data)
    ind = int(torch.argmax(y_pred, -1))
    recommend = recset[recset["track_genre"] == genre_dict[ind]].sample(n = 10)
    return recommend[['artists', 'album_name', 'track_name', 'track_genre', 'track_id']]
    
```

```python
#test upload_recommend
upload_recommend("data/test/test.csv")
```

```python
#recommendation based on user's history
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

#input parameter "history" is a dataframe that stores artists, album_name, track_name, and genre
#latest history appears at the tail
def history_recommend(history):
    #when history is empty, randomly present 10 songs
    if history.empty:
        recommend = recset.sample(n=10, replace=False)
        return recommend[['artists', 'album_name', 'track_name', 'track_genre', 'track_id']]
    
    songs = recset.copy()
    songs = songs.drop_duplicates(subset = ["track_id"])
    label_encoder = LabelEncoder()
    songs.iloc[:, -1] = label_encoder.fit_transform(songs.iloc[:, -1])
    songs_num = songs.drop(songs.columns[[6, 11]], axis = 1)
    scaler = MinMaxScaler()
    songs_num[songs_num.columns[4:]] = scaler.fit_transform(songs_num[songs_num.columns[4:]])
    songs_num = pd.DataFrame(songs_num)

    df_history = pd.merge(history, songs_num, on=['track_id'])
    df_history = df_history.drop(df_history.columns[0:8], axis = 1)
    songs_num = songs_num.drop(songs_num.columns[0:4], axis = 1)

    weights = list(range(1, len(df_history) + 1))[::-1]

    weighted_history = df_history.mul(weights, axis=0)
    weighted_avg_history = weighted_history.sum(axis=0) / sum(weights)

    #10 NearestNeighbors
    knn = NearestNeighbors(n_neighbors=10)
    knn.fit(songs_num.values)
    weighted_avg_vector = weighted_avg_history.values.reshape(1, -1)
    _, indices = knn.kneighbors(weighted_avg_vector)

    # Get the indices of the neighbors
    recommended_songs_indices = indices[0]
    recommended_songs = songs.iloc[recommended_songs_indices]
    recommended_songs.iloc[:, -1] = label_encoder.inverse_transform(recommended_songs.iloc[:, -1])

    return recommended_songs[['artists', 'album_name', 'track_name', 'track_genre', 'track_id']]    
```
